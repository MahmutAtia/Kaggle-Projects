## Prepare the documents for TF-IDF

```python
docs_dict= {}
indx = 0
for film in subset.iloc[:,1:].itertuples():
  doc = " "
  for i in film[1:]:
    doc +=( " " + str(i).replace("|" , " "))
  docs_dict[sub_set["title"][indx]]= doc
  indx+=1

```

```python
# checking the result
for key in list(docs_dict.keys())[0:3]:
  print(f"""
  Film Name : {key}
  The Text : 
  {docs_dict[key]}
  """)
```

```python
# very good now we can build the TF-IDF Vectorizer
tf_idf = TfidfVectorizer()
all_films = tf_idf.fit_transform(list(docs_dict.values()))
```

```python
#trying 
chosen_film = 'Jai Ho (film)'
film_doc = docs_dict[chosen_film]
trans_doc = tf_idf.transform([film_doc])
simi_array = cosine_similarity(trans_doc, all_films).flatten()
```

```python
simi_df = pd.DataFrame({
    "title":list(docs_dict.keys()) ,
    "similarity":simi_array 
                        })
```

```python
simi_df_sorted = simi_df.sort_values( by=["similarity"], ascending=False)
```

## Bulding a Function to recommend the most similar 5 films

```python
def most_similar(film): 

  chosen_film = film
  film_doc = docs_dict[chosen_film]
  trans_doc = tf_idf.transform([film_doc])
  simi_array = cosine_similarity(trans_doc, all_films).flatten()
  simi_df = pd.DataFrame({
    "title":list(docs_dict.keys()) ,
    "similarity":simi_array 
                        })
  simi_df_sorted = simi_df.sort_values( by=["similarity"], ascending=False)
  
  return simi_df_sorted["title"][1:6]


```
